```python
import pandas as pd
import numpy as np
```

```python
#read the file
df = pd.read_csv("topyoutube.csv")
df
```

```python
#Multiplying the subscribers by 100 million
df['100M'] = df['100M'].transform(lambda x: x * 100*10**6)
df
```

```python
#Filling NaN with string
df = df.fillna('Não informado')
df
```

```python
#Rounding the Avg variable
df['Avg'] = df['Avg'].round(2)
df
```

```python
#Showing only the top 10 artists
df.head(10)
```

```python
#Showing the first 100 artists sorted by the name
df.sort_values('Artist').head(100)
```

```python
#Saving a new file with the modified data
df.to_csv('top_youtubers_modified.csv')
```
