```python
x = ['bananas', 'maçãs', 'peras', 'uvas', 'laranjas']
y = ''
while y != '999':
  y = input('Qual fruta você quer checar?')
  if y == '999': print('Finalizando lista de compras')
  elif y in x:
    print('Já está na lista')
  else:
    x = x + [y]
    print('Fruta adicionada')
print('\n Lista de compras atualizada')
print(x)
```
